# pandas series

```python
import numpy as np
import pandas as pd
```

```python
q = [1,2,3,4,5]
w = [10,20,30,40,50]

pd.Series(w,q)
```

```python
dic = {1:20,2:40,3:60,4:80}
pd.Series(dic)
```

```python
a = np.arange(0,100)
b = np.arange(1,101)
pd.Series(b,a)
```

```python
arr=np.arange(0,99,9)
pd.Series(arr)
```

```python
w = [6,7,8,9,10,11]
q.extend(w)
q
```

```python
pd.Series(arr,q)
```

```python
sec1 = pd.Series([1,2,3,4,5],index=['usa','russia','india','china','japan'])
sec1
```

```python
sec2 = pd.Series([6,7,8,9,10],index=['usa','india','russia','china','japan'])
sec2
```

```python
sec1 + sec2
```

# missing data

```python
dff = pd.DataFrame({'a':[1,2,np.nan],
                    'b':[8,np.nan,np.nan],
                    'c':[4,5,6]})
dff
```

```python
dff.dropna()
```

```python
dff.dropna(axis=1)
```

```python
dff.fillna(value=10)
```

```python
dff.fillna(value=dff['a'].mean())
```
